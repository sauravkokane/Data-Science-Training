```python
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set(style="white", color_codes=True, rc={'figure.figsize':(18,11)})
```

Reading data from file

```python
enrolment_2018 = pd.read_csv('/content/drive/MyDrive/Government Datasets/enrolment_age_2018_19.csv')
```

```python
enrolment_2018.head()
```

```python
enrolment_2018.shape
```

```python
enrolment_2018.drop(["state_cd", "district_cd"], axis=1).describe().T
```

```python
classwise_enrollment = enrolment_2018.copy()

for i in range(1,13):
  classwise_enrollment[f"class_{i}"] = enrolment_2018[f"class_{i}_boys"] + enrolment_2018[f"class_{i}_girls"]
  classwise_enrollment.drop([f"class_{i}_boys", f"class_{i}_girls"], axis=1, inplace=True)
classwise_enrollment.head()
```

```python
classwise_enrollment.drop(["state_cd", "district_cd"], axis=1).describe().T
```

```python
# printing mean of each class enrollment
for i in range(1,13):
  print(f"There are average {classwise_enrollment[f'class_{i}'].mean()} students enrolled in class {i}")
```

```python
# printing median of each class enrollment
for i in range(1,13):
  print(f"Median of students enrolled in class {i} is {classwise_enrollment[f'class_{i}'].median()} students")
```

```python
# printing variance of each class enrollment
for i in range(1,13):
  print(f"Variance of students enrolled in class {i} is {classwise_enrollment[f'class_{i}'].var()}")
```

```python
# printing variance of each class enrollment
# Data is dispersed in a wide range
for i in range(1,13):
  print(f"Standard deviation of students enrolled in class {i} is {classwise_enrollment[f'class_{i}'].std()}")
```

```python
for i in range(1,13):
  print(f"range of number of students enrolled in class {i} is {classwise_enrollment[f'class_{i}'].min()} - {classwise_enrollment[f'class_{i}'].max()}")
```

```python
statewise_enrolment = enrolment_2018.drop(["age", "district_cd", "district_name"], axis=1).groupby(["state_name",  "state_cd", "ac_year"]).sum().reset_index()

statewise_enrolment.head()
```

```python
fig, ax = plt.subplots()
ax = sns.histplot(data=enrolment_2018, x="age", kde=True, bins=50)
mean = classwise_enrollment["age"].mean()
median = classwise_enrollment["age"].median()
ax.axvline(median, color='g', linestyle='-', label='Median')
ax.axvline(mean, color='r', linestyle='--', label='Mean')
plt.show()
```

```python
fig, ax = plt.subplots()
ax = sns.histplot(data=classwise_enrollment, x="class_1", kde=True, bins=50)
mean = classwise_enrollment["class_1"].mean()
median = classwise_enrollment["class_1"].median()
ax.axvline(median, color='g', linestyle='-', label='Median')
ax.axvline(mean, color='r', linestyle='--', label='Mean')
plt.show()
```

```python
fig, ax = plt.subplots()
ax = sns.histplot(data=classwise_enrollment, x="class_2", kde=True, bins=50)
mean = classwise_enrollment["class_2"].mean()
median = classwise_enrollment["class_2"].median()
ax.axvline(median, color='g', linestyle='-', label='Median')
ax.axvline(mean, color='r', linestyle='--', label='Mean')
plt.show()
```

```python
boxplot = sns.boxplot(x="state_name", y="class_1", data=classwise_enrollment)
boxplot.set_xticklabels(boxplot.get_xticklabels(), rotation=45, ha="right")
plt.show()
```

```python
data = enrolment_2018.state_name.value_counts()
sns.barplot(x=data.index, y=data)
plt.xticks(rotation=90, ha="right")
plt.show()
```
