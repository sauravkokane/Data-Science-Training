# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
# connect to drive
from google.colab import drive
drive.mount('/content/drive')
```

```
dataset_path = "/content/drive/MyDrive/Machine Learning A-Z/Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv"
```

```
dataset = pd.read_csv(dataset_path)
dataset
```

```
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
X, y
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
pf2 = PolynomialFeatures(degree=2)
X_pf2 = pf2.fit_transform(X)
linear_regressor2 = LinearRegression()
linear_regressor2.fit(X_pf2, y)
```

```
pf4 = PolynomialFeatures(degree=4)
X_pf4 = pf4.fit_transform(X)
linear_regressor4 = LinearRegression()
linear_regressor4.fit(X_pf4, y)
```

## Visualising the Linear Regression results

```
plt.scatter(X, y, color='red')
plt.plot(X, linear_regressor.predict(X), color='blue', linestyle='--')
plt.title("Truth / Bluff (Linear Regression)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.scatter(X, y, color='red')
plt.plot(X, linear_regressor2.predict(pf2.fit_transform(X)), color='blue', linestyle='--')
plt.title("Truth / Bluff (Polynomial Regression of order 2)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```
plt.scatter(X, y, color='red')
plt.plot(X, linear_regressor4.predict(pf4.fit_transform(X)), color='blue', linestyle='-')
plt.title("Truth / Bluff (Polynomial Regression of order 2)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(X.min(), X.max(), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
```

```
plt.scatter(X, y, color='red')
plt.plot(X_grid, linear_regressor4.predict(pf4.fit_transform(X_grid)), color='blue', linestyle='-')
plt.title("Truth / Bluff (Polynomial Regression of order 4)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

## Predicting a new result with Linear Regression

```
y_check = linear_regressor.predict([[6.5]])
y_check
```

## Predicting a new result with Polynomial Regression

```
y_check = linear_regressor2.predict(pf2.transform([[6.5]]))
y_check
```

```
y_check = linear_regressor4.predict(pf4.transform([[6.5]]))
y_check
```
